En este Script se entrena un modelo con los conjuntos de datos train y test anteriores, y se guardan gr√°ficos sobre dicho proceso, para analizar el rendimiento.

Primero, se deben importar las librer√≠as necesarias.

```python
import os
from ultralytics import YOLO
import torch
import matplotlib.pyplot as plt
from PIL import Image
```

Antes de comenzar, se debe comprobar si cuda est√° disponible en el dispositivo. En caso de poder usar cuda, el entrenamiento se realizar√° usando la GPU, y en caso contrario, la CPU (tarda m√°s).

```python
# Comprobamos si Cuda est√° disponible, para poder entrenar en GPU
use_cuda = torch.cuda.is_available()

print(f"Version de PyTorch: {torch.__version__}\nSi no tiene el formato 2.X.X+cu1XX, entonces no puedes utilizar la aceleraci√≥n de GPU")
print(f"¬øCUDA disponible? {torch.cuda.is_available()}")

if use_cuda:
    print(f"nombre de GPU: {torch.cuda.get_device_name(0)}")
else:
    print(f"Debes instalar la versi√≥n de CUDA correspondiente desde: https://pytorch.org/get-started/locally/")
    print("Si tienes NVidia, puedes comprobar tu versi√≥n de cuda escribiendo en la terminal el comando:\nnvidia-smi")
```

A continuaci√≥n, se definen los directorios de donde extraer los datos, y donde guardar los modelos entrenados

```python
CWD = os.getcwd()
DATASET_PATH = os.path.join(CWD, "dataset")

models_to_train = ["yolo11n-cls.pt", "yolo11s-cls.pt", "yolo11m-cls.pt"]

EPOCHS = 30         # N√∫mero de pasadas por el dataset
IMGSZ = 224         # Tama√±o de imagen
BATCH_SIZE = 16     # Puedes ponerlo a 64 si tienes una buena gr√°fica para tardar menos
PATIENCE = 10       # Detener si no mejora tras 10 √©pocas
PROJECT_NAME = "runs" # Carpeta donde se guardar√°n los resultados
```

Por √∫ltimo, se va a entrenar tres modelos distintos de Yolo 11, los modelos nano (n), small (s) y medium (m).

```python
device_to_use = 'cuda' if torch.cuda.is_available() else 'cpu'

for model_name in models_to_train:
    run_name = f'train_{model_name}'
    
    print(f"\n\n-------------------Modelo: {model_name}-------------------\n")
    
    model = YOLO(model_name) # Cargamos el modelo
    
    model.train(
        data=DATASET_PATH,
        epochs=EPOCHS,
        imgsz=IMGSZ,
        batch=BATCH_SIZE,
        patience=PATIENCE, 
        project=PROJECT_NAME,
        name=run_name,
        device=device_to_use,
        verbose=True
    )
```
