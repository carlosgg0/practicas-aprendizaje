En este Script se entrena un modelo con los conjuntos de datos train y test anteriores, y se guardan gráficos sobre dicho proceso, para analizar el rendimiento.

Primero, se deben importar las librerías necesarias.

```python
import os
from ultralytics import YOLO
import torch
import matplotlib.pyplot as plt
from PIL import Image
```

Antes de comenzar, se debe comprobar si cuda está disponible en el dispositivo. En caso de poder usar cuda, el entrenamiento se realizará usando la GPU, y en caso contrario, la CPU (tarda más).

```python
# Comprobamos si Cuda está disponible, para poder entrenar en GPU
use_cuda = torch.cuda.is_available()

print(f"Version de PyTorch: {torch.__version__}\nSi no tiene el formato 2.X.X+cu1XX, entonces no puedes utilizar la aceleración de GPU")
print(f"¿CUDA disponible? {torch.cuda.is_available()}")

if use_cuda:
    print(f"nombre de GPU: {torch.cuda.get_device_name(0)}")
else:
    print(f"Debes instalar la versión de CUDA correspondiente desde: https://pytorch.org/get-started/locally/")
```

A continuación, se definen los directorios de donde extraer los datos, y donde guardar los modelos entrenados

```python
CWD = os.getcwd()
DATASET_PATH = os.path.join(CWD, "dataset")

models_to_train = ["yolo11n-cls.pt", "yolo11s-cls.pt", "yolo11m-cls.pt"]

EPOCHS = 30         # Número de pasadas por el dataset
IMGSZ = 224         # Tamaño de imagen
BATCH_SIZE = 16  
PATIENCE = 10       # Detener si no mejora tras 10 épocas
PROJECT_NAME = "runs" # Carpeta donde se guardarán los resultados
```

Por último, se va a entrenar tres modelos distintos de Yolo 11, los modelos nano (n), small (s) y medium (m).

```python
device_to_use = 'cuda' if torch.cuda.is_available() else 'cpu'

for model_name in models_to_train:
    run_name = f'train_{model_name}'
    
    print(f"\n\n-------------------Modelo: {model_name}-------------------\n")
    
    model = YOLO(model_name) # Cargamos el modelo
    
    model.train(
        data=DATASET_PATH,
        epochs=EPOCHS,
        imgsz=IMGSZ,
        batch=BATCH_SIZE,
        patience=PATIENCE, 
        project=PROJECT_NAME,
        name=run_name,
        device=device_to_use,
        verbose=True
    )
```
